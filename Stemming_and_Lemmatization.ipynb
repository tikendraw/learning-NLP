# Stemming and Lemmatization

>artificial , artificially --> arificia  </br>

**Lemmatization** : not stemming.

```python
from IPython.display import clear_output
import nltk
nltk.download('all')
clear_output(wait=True)
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, RegexpStemmer, SnowballStemmer, LancasterStemmer, WordNetLemmatizer
```

```python
paragraph = """Machine learning (ML) is a field of inquiry devoted to understanding and building methods that 'learn', that is, methods that leverage data to improve performance
             on some set of tasks.[1] It is seen as a part of artificial intelligence. Machine learning algorithms build a model based on sample data, known as training data, in
              order to make predictions or decisions without being explicitly programmed to do so.[2] Machine learning algorithms are used in a wide variety of applications, such
               as in medicine, email filtering, speech recognition, and computer vision, where it is difficult or unfeasible to develop conventional algorithms to perform the needed 
               tasks.[3]A subset of machine learning is closely related to computational statistics, which focuses on making predictions using computers, but not all machine learning
                is statistical learning. The study of mathematical optimization delivers methods, theory and application domains to the field of machine learning. Data mining is a
                 related field of study, focusing on exploratory data analysis through unsupervised learning.[5][6] Some implementations of machine learning use data and neural networks
                  in a way that mimics the working of a biological brain.[7][8] In its application across business problems, machine learning is also referred to as predictive analytics."""
```

### Turning our Paragraph into sentence

```python
# TOKENIZE THE SENTENCE
sentences = nltk.sent_tokenize(paragraph)
```

```python
print('No. of senetences: ',len(sentences))
print(sentences)
```

### Break the sentences into words and Use different Stemming technique

```python
pstem = PorterStemmer()
# rstem = RegexpStemmer()
sstem = SnowballStemmer(language='english')
lstem = LancasterStemmer()
wlemma = WordNetLemmatizer()
```

```python
actual_word = []
pstem_words = []
sstem_words = []
lstem_words = []
wlemma_words = []
```

```python
for _, sentence in enumerate(sentences):
    words = nltk.word_tokenize(sentence)
    for word in words:
        if word not in set(stopwords.words('english')):
            actual_word.append(word)
            pp = pstem.stem(word)
            pstem_words.append(pp)

            ss = sstem.stem(word)
            sstem_words.append(ss)

            ll = lstem.stem(word)
            lstem_words.append(ll)

            ww = wlemma.lemmatize(word)
            wlemma_words.append(ww)
```

```python
import pandas as pd
```

```python
stems = pd.DataFrame({'Actual_words':actual_word,
                     'PorterStemmer': pstem_words, 
                      'SnowballStemmer':sstem_words, 
                      'LancasterStemmer':lstem_words,
                      'Lemmatizer': wlemma_words})
```

```python
stems.sample(20)
```

```python
stems[stems.PorterStemmer==stems.Lemmatizer].sample(10)
```
